```
install.packages("ggplot2")
install.packages("SportsAnalytics")
library(SportsAnalytics)
library(ggplot2)
```

```
NBA1011 <- fetch_NBAPlayerStatistics("10-11")
```

```
str(NBA1011)
```

```
# 散布圖 隊伍
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1011, color = Team)
```

```
# 散布圖 位置
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1011, color = Position)
```

```
# 長條圖 位置
qplot(TotalPoints, data = NBA1011, fill = Position)
```

```
# 長條圖 隊伍
qplot(TotalPoints, data = NBA1011, fill = Team)
```

```
# 盒狀圖 FieldGoalsAttempted TotalPoints
ggplot(NBA1011, aes(x = FieldGoalsAttempted, y= TotalPoints)) +geom_boxplot()
```

```
# 盒狀圖 GamesPlayed TotalPoints
ggplot(NBA1011, aes(x = GamesPlayed, y= TotalPoints)) +geom_boxplot()
```

```
# 折線圖  FieldGoalsAttempted TotalPoints
ggplot(NBA1011, aes(x = FieldGoalsAttempted, y= TotalPoints))+ geom_line()
```

```
# 折線圖  GamesPlayed TotalPoints
ggplot(NBA1011, aes(x = GamesPlayed, y= TotalPoints))+ geom_line()
```
